9. Crea una nueva columna price_category que clasifique:

* 'económico' si price < 10,000

* 'estándar' si 10,000 ≤ price ≤ 20,000

* 'premium' si price > 20,000

Luego, cuenta cuántos autos hay en cada categoría.

9.1 Crear la columna price_category

```python
def clasificar_precio(precio):
    if precio < 10000:
        return 'económico'
    elif 10000 <= precio <= 20000:
        return 'estándar'
    else:
        return 'premium'

df['price_category'] = df['price'].apply(clasificar_precio)
```

16. Usando el método del rango intercuartil (IQR), detecta outliers en la columna horsepower. ¿Cuántos outliers encuentras?

16.1 Asegurar que horsepower sea numérico

```python
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
```

```python
Q1 = df['horsepower'].quantile(0.25)
Q3 = df['horsepower'].quantile(0.75)

IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR
```

```python
outliers = df[(df['horsepower'] < limite_inferior) | (df['horsepower'] > limite_superior)]

print(len(outliers))

outliers[['horsepower']]
```
