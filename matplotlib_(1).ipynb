>
    COM       ABR

    cyan ---- c
    magenta - m
    yellow -- y  
    black --- k
    white --- w

De forma similar, podemos pasar el estilo de la línea de forma completa o abreviada:
>
    COM        ABR      DESCRIPCIÓN

* **Estilo orientado a objetos**

Si bien la mayoría de las funciones en `plt` se traducen directamente a métodos `ax` (como `plt.plot()` → `ax.plot()`, `plt.legend()` → `ax.legend()`, etc.), este no es el caso para todos los comandos. En particular, las funciones para establecer límites, etiquetas y títulos se modifican ligeramente. Para realizar la transición entre las funciones de estilo MATLAB y los métodos orientados a objetos, realizamos los siguientes cambios (Puede consultar todos los atributos y métodos de la clase `axes` en este [link](https://matplotlib.org/3.1.1/api/axes_api.html#array)):

* `plt.plot()`   → `ax.plot()`
* `plt.legend()` → `ax.legend()`
* `plt.axis()`   → `ax.axis()`

* `plt.xlabel()` → `ax.set_xlabel()`
* `plt.ylabel()` → `ax.set_ylabel()`
* `plt.xlim()`   → `ax.set_xlim()`
* `plt.ylim()`   → `ax.set_ylim()`
* `plt.title()`  → `ax.set_title()`

## Tipos de graficos

**¿Qué es un histograma?**

Es un gráfico que muestra cómo se distribuyen
los valores de una variable numérica
dividiéndolos en "rangos" o "intervalos", en
lugar de mostrar cada valor individual, agrupa
los datos en bloques y cuenta cuántos valores
caen dentro de cada rango.

**¿Cuándo usarlo?**

Cuando necesitas ver la forma de una variable
numérica: si es simétrica, sesgada,
concentrada, dispersa, etc.

## ¿Qué es un boxplot?
Un boxplot (o diagrama de caja) es un gráfico
estadístico que resume una variable numérica
usando cinco valores clave:
* **Mínimo** (sin contar outliers)
* **Primer cuartil (Q1)** – 25% más bajo
* **Mediana (Q2)** – valor central
* **Tercer cuartil (Q3)** – 75% más bajo
* **Máximo** (sin contar outliers)

¿Qué permite ver?
* Si los datos están concentrados o dispersos
* Si hay valores extremos (outliers)
* Si los datos están simétricos o sesgados
* La posición de la mediana respecto a los
cuartile

## ¿Qué es un gráfico de barras?
Es un gráfico que muestra la frecuencia o
cantidad asociada a cada categoría de una
variable categórica, cada barra representa una
categoría, y su altura representa el número de
veces que aparece en los datos.

**¿Cuándo usarlo?**

Cuando tienes una variable categórica y quieres
visualizar su comportamiento.

**¿Qué permite ver?**

* La categoría que más aparece
* Categoría que tienen pocos datos
* Si existen clases desbalanceadas
* Si alguna categoría domina el conjunto de
datos

## ¿Qué es un pie chart?
Es un gráfico que representa proporciones de
un total, donde cada categoría se muestra
como una “rebanada” de un círculo, todo el
gráfico suma el 100%.

**¿Cuándo usarlo?**
* Para mostrar la proporción relativa de cada
categoría.
* No recomendado cuando hay muchas
categorías.
**¿Qué permite ver?**
* Qué categoría tiene la mayor proporción
* Si una categoría domina claramente
* Cómo se reparte el total entre las partes

## ¿Qué es un violin plot?
Un violin plot muestra la distribución de una
variable numérica, también incluye una curva
de densidad que representa la frecuencia de
los valores. La forma del “violín” indica dónde
se concentran más los datos.

**¿Cuándo usarlo?**
* Cuando el boxplot es insuficiente para
entender la forma de la distribución.
* Ideal para comparar una variable numérica
entre diferentes categorías

**¿Qué permite ver?**
* Si los datos están concentrados o dispersos
* Si hay una o varias modas (picos de densidad)
* La asimetría en la distribución
* Outliers y mediana



## ¿Qué es un scatterplot?

Un scatterplot muestra una nube de puntos en
un plano, donde:
* El eje X representa una variable numérica.
* El eje Y representa otra variable numérica.
* Cada punto es una observación del conjunto
de datos.
Sirve para detectar relaciones o patrones entre
dos variables.

**¿Qué permite ver?**
* Si dos variables se relacionan entre sí
* Qué tan fuerte o débil es esa relación
* Si hay grupos separados
* Si hay valores atípicos que no siguen la
tendencia

## ¿Qué es un line plot?
Es un gráfico que conecta puntos de datos con
líneas, para mostrar cómo cambia una variable
numérica en función de otra variable,
normalmente tiempo (fecha, año, trimestre,
etc.).

**¿Cuándo usarlo?**
* Analizar tendencias
* Ver fluctuaciones o estacionalidades
* Comparar antes y después de un evento

**¿Qué permite ver?**
* Si hay una tendencia creciente o decreciente
* Picos o caídas abruptas
* Estabilidad o variabilidad en el tiempo
* Posibles ciclos o patrones repetitivos

## ¿Qué es un heatmap de correlación?
Es una matriz visual donde cada celda
representa el coeficiente de correlación entre
dos variables numéricas, el color indica la
fuerza y dirección de la relación:
*  +1: correlación positiva perfecta (ambas
suben o bajan juntas)
*  0: sin relación lineal
*  −1: correlación negativa perfecta (una sube, la
otra baja)
**¿Qué permite ver?**
* Si hay variables que aumentan o disminuyen
juntas
* Posibles multicolinealidades (útil para
regresiones)
* Dónde hay relaciones débiles o irrelevantes
* Evitar redundancias en modelado
* Identificar relaciones inesperadas

## ¿Qué es un pairplot?
Genera una cuadrícula de gráficos con todas
las combinaciones posibles entre variables
numéricas de un DataFrame:
* Las diagonales muestran la distribución
univariada de cada variable (histograma o
kde).
* Las celdas fuera de la diagonal muestran
gráficos de dispersión entre pares de
variables.
* Puede incluir colores por categoría si se añade
una variable categórica (ideal para clasificar o
segmentar).

**¿Qué permite ver?**
*  Relaciones lineales o no lineales entre pares
de variables
* Variables que se comportan de forma similar
* Agrupamientos de datos (especialmente si se
usa hue)
* Outliers o valores extremos que se repiten
entre pares


**¿Qué es FacetGrid?**
Es una técnica que permite dividir un gráfico
en múltiples subgráficos, cada uno
correspondiente a un subgrupo de una variable
categórica.
Con FacetGrid, puedes ver cómo cambia la
relación o distribución de variables según
diferentes categorías.

**¿Qué permite ver?**
* Cambios de patrón entre categorías.
* Similitudes o diferencias visuales entre
subgrupos.
* Agrupamientos, outliers o relaciones
específicas de un grupo.
